```python
import pandas as pd
import numpy as np
df=pd.read_csv('night_out.csv')
print(df)
print("----------------------------------------------------------------------------------------")
print("-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*")
print("----------------------------------------------------------------------------------------")
x=df.iloc[:,0:3]
x=pd.get_dummies(x,drop_first=True)
y=df.iloc[:,-1]
from sklearn.tree import DecisionTreeClassifier,plot_tree
d_tree=DecisionTreeClassifier(criterion='entropy',max_depth=5)
d_tree.fit(x,y)
y_pred=d_tree.predict(x)
from sklearn.metrics import accuracy_score
print(accuracy_score(y,y_pred),':acc_dt')
print("----------------------------------------------------------------------------------------")
print("Prediction for DT:",d_tree.predict(np.array([True,False,True,True]).reshape(1,-1)))
print("----------------------------------------------------------------------------------------")
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=3)
model=knn.fit(x,y)
pred=model.predict(x)
print(accuracy_score(y,pred),':acc_knn')
print("----------------------------------------------------------------------------------------")
print("Prediction for Knn:",knn.predict(np.array([True,False,True,True]).reshape(1,-1)))
print("----------------------------------------------------------------------------------------")
```

```python
import pandas as pd
df=pd.read_csv('diabetes.csv')
print(df)
print("----------------------------------------------------------------------------------------")
x=df.iloc[:,0:8]
x=pd.get_dummies(x,drop_first=True)
y=df.iloc[:,-1]
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier(n_estimators=100,random_state=42,max_depth=6)
rf.fit(x_train,y_train)
y_pred=rf.predict(x_test)
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
print('cm:',confusion_matrix(y_test,y_pred))
print(accuracy_score(y_test,y_pred),':acc')
print(classification_report(y_test,y_pred))


```

```python
import pandas as pd
df=pd.read_csv('glass.csv')
print(df)
print("----------------------------------------------------------------------------------------")
x=df.iloc[:,0:9]
x=pd.get_dummies(x,drop_first=True)
y=df.iloc[:,-1]
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier(n_estimators=100,random_state=42,max_depth=10)
rf.fit(x_train,y_train)
y_pred=rf.predict(x_test)
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
print("----------------------------------------------------------------------------------------")
print('cm:',confusion_matrix(y_test,y_pred))
print("----------------------------------------------------------------------------------------")
print(accuracy_score(y_test,y_pred),':acc')
print("----------------------------------------------------------------------------------------")
print(classification_report(y_test,y_pred))
print("----------------------------------------------------------------------------------------")
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=15)
model=knn.fit(x_test,y_test)
pred=model.predict(x_test)
print(accuracy_score(y_test,pred),':acc_knn')
print("----------------------------------------------------------------------------------------")
```
